```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

## Cleaning and preprocessing table

```python
df2 = pd.read_csv('bsc28.csv')
df2.head()
```

```python
df2['Cell Name'] = df2['Cell Name'].str[-1:]
df2.head()
```

```python
df2 = df2.groupby(by='Cell ID').agg({'Cell Name':lambda x: list(x)})
df2['Cell Name'] = [','.join(map(str, l)) for l in df2['Cell Name']]
df2.head()
```

## Load main table

```python
df = pd.read_csv('Anuradhapura.csv')
df = df.drop('Cell-on Air', 1)
df.head()
```

## Generate new dataframe

```python
newDf = df.merge(df2, left_on='Site ID', right_on='Cell ID')
newDf.head()
```

```python
newDf.rename(columns={"Cell Name" : "Cell-on Air"})
```

```python
newDf.to_csv('out.csv')
```
